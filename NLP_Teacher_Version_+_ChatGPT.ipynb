```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Define the Wikipedia page URL
url = 'https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table'

# Use requests to get the page HTML
response = requests.get(url)

# Use BeautifulSoup to parse the HTML and find the Olympic games table
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table', {'class': 'wikitable sortable plainrowheaders'})

# Use pandas to read the HTML table into a dataframe
df = pd.read_html(str(table))[0]

# Clean up the dataframe by renaming columns and removing unnecessary columns and rows
df.columns = ['Rank', 'Team/NOC', 'Total', 'Gold', 'Silver', 'Bronze', 'Combined Total']
df = df.iloc[:-1]
df = df.drop(['Combined Total'], axis=1)
df = df.reset_index(drop=True)

# Print the dataframe
print(df)
```
